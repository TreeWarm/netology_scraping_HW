```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
res = requests.get('https://habr.com/ru/')
res.status_code
```

```python
def get_habr_blog(query):
    habr_blog = pd.DataFrame()
    for q in query:
        URL = 'https://habr.com/ru/search/'
        params = {'q': q}
        req = requests.get(URL, params=params)
        time.sleep(0.3)
        soup = BeautifulSoup(req.text)
        posts = set(soup.find_all('article', class_='tm-articles-list__item'))
#         –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞—é, —á—Ç–æ set –∏–∑–±–∞–≤–∏—Ç –æ—Ç –ø–æ–≤—Ç–æ—Ä–æ–≤ –ø–æ—Å—Ç–æ–≤
        
        for el in posts:
            title = el.find('a', 'tm-article-snippet__title-link').text
            link = el.find('a', 'tm-article-snippet__title-link').get('href')
            date = el.find('time').get('title')
            likes = el.find('div', 'tm-votes-meter').text
                        
            row = {'date': date, 'title': title, 'link': f'https://habr.com'+link, 'likes': likes}
            habr_blog = pd.concat([habr_blog, pd.DataFrame([row])])
            
    return habr_blog.reset_index(drop=True)
res = get_habr_blog(['vst', '—Å–∏–Ω—Ç–µ–∑–∞—Ç–æ—Ä—ã'])
res
```
